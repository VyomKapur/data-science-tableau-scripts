```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from os import listdir
from os.path import isfile, join
```

```python
fileset1 = [f for f in listdir('./archive') if isfile(join('./archive', f))]
fileset1
```

```python
fileset2 = [f for f in listdir('./archive/crime') if isfile(join('./archive/crime', f))]
fileset2
```

```python
dfset = []
for i in fileset2:
    try:
        dfset.append(pd.read_csv('./archive/crime/'+i))
    except:
        print("error", i)
```

```python
df1 = dfset[0]
df2 = dfset[1]
df3 = dfset[2]
df1.head()
```
